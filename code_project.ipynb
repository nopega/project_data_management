```python
import pandas as pd 
import os
```

```python
folder_path = 'C:\\Users\\pongk\\Documents\\projed_code_academy\\JC_DATA'
list_file_name = []
for file in os.listdir(folder_path):
    if file.startswith('JC'):
        list_file_name.append(file)
```

```python
dataframe =[]
for file_name in list_file_name:
    #print(file_name)
    data = pd.read_csv("JC_DATA\\"+file_name)
    dataframe.append(data)

dataframe = pd.concat(dataframe, ignore_index=True)
```
